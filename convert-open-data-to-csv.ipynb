```python
import json
import pandas as pd
```

```python
with open("data/2023-01-17.geojson") as fp:
    data = json.load(fp)
```

```python
list_fuels = ["Gazole", "SP95", "E10", "SP98", "E85", "GPLc"]
```

```python
array_stations = []
for f in data["features"]:
    dict_station = {}
    # Initialisation des colonnes à None
    dict_station["id_pdv"] = f["properties"]["id"]
    for lf in list_fuels:
        dict_station["prix_" + lf] = None
        dict_station["date_prix_" + lf] = None
        dict_station["rupture_" + lf] = None
        dict_station["date_rupture_" + lf] = None
    dict_station["fermeture_station"] = None
    dict_station["date_debut_fermeture_station"] = None
    dict_station["date_fin_fermeture_station"] = None
    # pour chaque carburants récupérer les infos de prix et de ruptures s'il y a lieu
    for lf in list_fuels:
        for item in f["properties"]["prix"]:
            if item["nom"] == lf:
                dict_station["prix_" + lf] = item["valeur"]
                dict_station["date_prix_" + lf] = item["maj"]
        for item in f["properties"]["ruptures"]:
            if item["nom"] == lf:
                dict_station["rupture_" + lf] = "R"
                dict_station["date_rupture_" + lf] = item["debut"]
    # pour chaque fermetures
    for ferm in f["properties"]["fermetures"]:
        dict_station["fermeture_station"] = ferm["type"]
        dict_station["date_debut_fermeture_station"] = ferm["debut"]
        dict_station["date_fin_fermeture_station"] = ferm["fin"]
    array_stations.append(dict_station)
```

```python
df_stations = pd.DataFrame(array_stations)
```

```python
df_stations.to_csv("open-data-carburants-2023-01-17.csv", index=False, sep=";")
```
